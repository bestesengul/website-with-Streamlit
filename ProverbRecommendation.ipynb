```python
#Ask
AskYazar = []
AskSoz = [soz for soz in Ask.split('\n') if soz != '']  #ozlu sozleri birbirinden ayirdik

for yazar in range(len(AskSoz)):
    AskYazar.append(AskSoz[yazar].split('#', 1)[1]) #her bir cumlenin sonunda yer alan ve # isaretinden sonra olan yazar ismini aldik

for soz in range(len(AskSoz)):
    AskSoz[soz] = AskSoz[soz].split('#', 1)[0]  #sozlerin icinde yer alan yazar isimlerini cikardik

#Mutluluk
MutlulukYazar = []

MutlulukSoz = [soz for soz in Mutluluk.split('\n') if soz != '']

for yazar in range(len(MutlulukSoz)):
    MutlulukYazar.append(MutlulukSoz[yazar].split('#', 1)[1])

for soz in range(len(MutlulukSoz)):
    MutlulukSoz[soz] = MutlulukSoz[soz].split('#', 1)[0]

#Basari
BasariYazar = []

BasariSoz = [soz for soz in Basari.split('\n') if soz != '']

for yazar in range(len(BasariSoz)):
    BasariYazar.append(BasariSoz[yazar].split('#', 1)[1])

for soz in range(len(BasariSoz)):
    BasariSoz[soz] = BasariSoz[soz].split('#', 1)[0]

#Hayat
HayatYazar = []

HayatSoz = [soz for soz in Hayat.split('\n') if soz != '']

for yazar in range(len(HayatSoz)):
    HayatYazar.append(HayatSoz[yazar].split('#', 1)[1])

for soz in range(len(HayatSoz)):
    HayatSoz[soz] = HayatSoz[soz].split('#', 1)[0]

sozler = AskSoz + MutlulukSoz + BasariSoz + HayatSoz
kategori = ['Ask'] * len(AskSoz) + ['Mutluluk'] * len(MutlulukSoz) + ['Basari'] * len(BasariSoz) + ['Hayat'] * len(HayatSoz)
yazarlar = AskYazar + MutlulukYazar + BasariYazar + HayatYazar

#jaccard fonksiyonu için kategoriler ve yazarlar list haline getirildi
kategoriler = ['ask', 'mutluluk', 'basari', 'hayat']
yazarTekIsimleri = []
yazarlarSet = list(set(yazarlar))
for i in range(len(yazarlarSet)):
    yazarTekIsimleri.append(yazarlarSet[i].lower())

veri = pd.DataFrame({"Sozler": sozler, "Kategori": kategori, "Yazar": yazarlar})
veri.head(4)
veri.to_csv('export_dataframe.csv', index = True, header=True)
```

```python
veri["temizSozler"] = veri['Sozler'].copy()

for i in range(len(veri)):
    veri["temizSozler"][i] = veri["temizSozler"][i].lower()
    veri["temizSozler"][i] = re.sub('[!@#’‘?.,\'$]', '', veri["temizSozler"][i])
```

```python
def jaccardSimilarity(list1, list2):
  return float(len(list(set(list1).intersection(list2)))) / len(list(set(list1).union(list2)))
```

```python
def recommendation_random_jaccard(girdi):
    jaccard_score_list = []

    if(girdi in yazarTekIsimleri): #eger yazar ismi girilirse o yazardan random bir soz dondurulur
        veriler = []
        for i in range(len(veri)):
            if(str(veri['Yazar'][i]).lower() == str(girdi)): 
                veriler.append(i)
        index = random.choice(veriler)
        recommendation = veri['Sozler'][index] + "- " + veri["Yazar"][index] 
        return recommendation

    if len(girdi.split()) > 1:
        for i in range(len(veri)):
            jaccard_score_list.append(jaccardSimilarity(veri["temizSozler"][i], girdi))        
            maxIndex = jaccard_score_list.index(max(jaccard_score_list))
            recommendation =veri['Sozler'][maxIndex] + "- " + veri['Yazar'][maxIndex]

        return recommendation

    else:
        if(girdi in kategoriler): #eger kategori ismi girerse o kategoriden random bir soz dondurulur
            veriler = []
            for i in range(len(veri)):
                if(str(veri['Kategori'][i]).lower() == str(girdi)): 
                    veriler.append(i)
            index = random.choice(veriler)
            recommendation = veri['Sozler'][index] + "- " + veri["Yazar"][index] 
            return recommendation

        else: # rastgele bir replik dondurulur
            index = veri[veri['Sozler'] == veri["Sozler"].sample(n=1).values[0]].index[0]
            recommendation = veri['Sozler'][index] + "- " + veri["Yazar"][index] 
            
        return recommendation
```

```python
kelimeVektorleri = KeyedVectors.load_word2vec_format('trmodel', binary=True)
def CreateWord2VecMatrices(veri, girdi):
        kelimeMatrisi = np.zeros((len(veri), 400))
        for i in range(len(veri)):
            word2vecToplam = np.zeros((1, 400))
            for j in veri.iloc[i]['temizSozler'].split(' '):
                try:
                    word2vecToplam += kelimeVektorleri[j]
                except: 
                    pass
            kelimeMatrisi[i] = np.divide(word2vecToplam, len(veri.iloc[i]['temizSozler'].split(' ')))

            girdiVektor = np.zeros((1, 400))
            for k in girdi.split(' '):
                try:
                    girdiVektor += kelimeVektorleri[k]
                except: 
                    pass
            girdiVektor = np.divide(girdiVektor, len(girdi.split(' ')))

        return kelimeMatrisi, girdiVektor
```

```python
def recommendation_wv(girdi):
        girdi = girdi.lower()

        if(girdi in yazarTekIsimleri): #eger yazar ismi girilirse o yazardan random bir soz dondurulur
            veriler = []
        
            for i in range(len(veri)):
                if(str(veri['Yazar'][i]).lower() == str(girdi)): 
                    veriler.append(i)
            index = random.choice(veriler)
            recommendation = veri['Sozler'][index] + "\n\n~ " + veri["Yazar"][index] 
                    
        else:
            kelimeMatrisi, girdiVektor = CreateWord2VecMatrices(veri, girdi)
            skorVerisi = []
            for i in range(len(veri)):
                skor = cosine_similarity(kelimeMatrisi[i].reshape(1,400), girdiVektor.reshape(1,400))
                skorVerisi.append(skor)
            recommendation = veri.iloc[skorVerisi.index(max(skorVerisi))].Sozler + "\n\n~ " + veri.iloc[skorVerisi.index(max(skorVerisi))].Yazar
            
        return recommendation
```
